```python
import pandas as pd

df = pd.read_csv('s&p500.csv')
df = df.set_index('Date')
df
```

```python
days_in_quarter = 64
num_quarters = df.shape[0] // days_in_quarter
num_days = num_quarters * days_in_quarter
df = df.iloc[:num_days]
df
```

```python
train_df = df.iloc[:int(num_quarters*0.8)]
```

```python
import os

def generate_correlations(corr_name, corr_func):
    # Global correlation
    os.makedirs(corr_name, exist_ok=True)
    global_corr = corr_func(train_df)
    np.savetxt(os.path.join(corr_name, 'global_corr.csv'), global_corr, delimiter=',')

    # Local correlations
    for q in range(num_quarters):
        local_df = df.iloc[q*days_in_quarter:(q+1)*days_in_quarter]
        local_corr = corr_func(local_df)
        np.savetxt(os.path.join(corr_name, f'local_corr_{q}.csv'), local_corr, delimiter=',')
```

# Mutual Information

```python
from sklearn.metrics import mutual_info_score
import numpy as np

def calculate_mutual_information(series1, series2, bins=5):
    """
    Calculate the mutual information between two series.
    series1, series2: Pandas Series
    bins: Number of bins for discretization
    """
    # Discretize the series
    data1 = pd.cut(series1, bins=bins, labels=False)
    data2 = pd.cut(series2, bins=bins, labels=False)

    # Compute mutual information
    return mutual_info_score(data1, data2)


def mutual_information_matrix(df, bins=5):
  """
    Calculate the mutual information matrix for all pairs of columns in the DataFrame.
    df: DataFrame with columns as assets and rows as daily log-return values.
    bins: Number of bins for discretization.
  """
  df_tickers = df.columns
  n = len(df.columns)
  mi_matrix = pd.DataFrame(np.zeros((n, n)), index=df_tickers, columns=df_tickers)
  for i in range(n):
    for j in range(i, n):
      mi = calculate_mutual_information(df.iloc[:, i], df.iloc[:, j], bins=bins)
      mi_matrix.iloc[i, j] = mi
      mi_matrix.iloc[j, i] = mi
  return mi_matrix
```

```python
import numpy as np
import torch
from scipy.stats import rankdata
from InfoNet.infer import load_model, infer
from tqdm import tqdm

config_path = "InfoNet/configs/config.yaml"
ckpt_path = "InfoNet/infonet_5000_32_1000-720--0.16.pt"
model = load_model(config_path, ckpt_path)
model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def mutual_information_matrix_infonet(df, batch_size=64):
    data = df.to_numpy()
    T = data.shape[0]
    N = data.shape[1]
    ranked_data = rankdata(data, axis=0)/T
    batch = np.zeros((batch_size, T, 2))
    batch_idx = 0
    batch_ijs = []
    MI = np.zeros((N, N))
    pbar = tqdm(total=(0.5*N*(N+1))//batch_size, desc="Calculating MI with InfoNet")
    for i in range(N):
        for j in range(i, N):
            if batch_idx == batch_size:
                with torch.no_grad():
                    res = model(torch.tensor(batch, dtype=torch.float32, device=device)).cpu().numpy()
                for (a, b), m in zip(batch_ijs, res):
                    MI[a, b] = m
                    MI[b, a] = m
                batch[:] = 0
                batch_idx = 0
                batch_ijs.clear()
                pbar.update(1)
            batch[batch_idx, :, :] = np.column_stack((ranked_data[:, i], ranked_data[:, j]))
            batch_ijs.append((i, j))
            batch_idx += 1
    pbar.close()

    if len(batch_ijs) > 0:
        with torch.no_grad():
            res = model(torch.tensor(batch[:len(batch_ijs)], dtype=torch.float32, device=device)).cpu().numpy()
        for (a, b), m in zip(batch_ijs, res):
            MI[a, b] = m
            MI[b, a] = m

    return MI
```

```python
generate_correlations('mi', mutual_information_matrix_infonet)
```

# Pearson's Correlation

```python
import numpy as np

def pearsons_matrix(df):
  data = df.to_numpy()
  c = np.corrcoef(data, rowvar=False)
  return c
```

```python
generate_correlations('pearsons', pearsons_matrix)
```
