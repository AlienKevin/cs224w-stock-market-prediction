```python
import pandas as pd

df = pd.read_csv('s&p500.csv')
df = df.set_index('Date')
df
```

```python
from sklearn.metrics import mutual_info_score
import numpy as np

def calculate_mutual_information(series1, series2, bins=5):
    """
    Calculate the mutual information between two series.
    series1, series2: Pandas Series
    bins: Number of bins for discretization
    """
    # Discretize the series
    data1 = pd.cut(series1, bins=bins, labels=False)
    data2 = pd.cut(series2, bins=bins, labels=False)

    # Compute mutual information
    return mutual_info_score(data1, data2)


def mutual_information_matrix(df, bins=5):
  """
    Calculate the mutual information matrix for all pairs of columns in the DataFrame.
    df: DataFrame with columns as assets and rows as daily log-return values.
    bins: Number of bins for discretization.
  """
  df_tickers = df.columns
  n = len(df.columns)
  mi_matrix = pd.DataFrame(np.zeros((n, n)), index=df_tickers, columns=df_tickers)
  for i in range(n):
    for j in range(i, n):
      mi = calculate_mutual_information(df.iloc[:, i], df.iloc[:, j], bins=bins)
      mi_matrix.iloc[i, j] = mi
      mi_matrix.iloc[j, i] = mi
  return mi_matrix
```
