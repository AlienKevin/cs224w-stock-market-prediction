```python
import pandas

df = pandas.read_csv('s&p500.csv')
df = df.set_index('Date')
df
```

```python
from sklearn.metrics import normalized_mutual_info_score
import numpy as np

T = len(df[df.columns[0]])
S = len(df.columns)

A = np.zeros((S,S))

for i1, s1 in enumerate(df.columns):
    for i2, s2 in enumerate(df.columns):
        assert(len(df[s1]) == T)
        assert(len(df[s2]) == T)
        A[i1, i2] = normalized_mutual_info_score(df[s1].tolist(), df[s2].tolist())

A.tofile('s&p500_pearson_correlation.csv', sep = ',')
```
