```python
import csv

sp500 = []

with open('sp500/S&P 500 Historical Components & Changes(08-17-2024).csv', 'r') as input:
    reader = csv.reader(input)
    latest = list(reader)[-1][-1]
    # yahoo finance uses - instead of . in symbol names
    sp500 = [name.replace('.', '-') for name in latest.split(',')]

print(len(sp500))
```

```python
import yfinance as yf

df = yf.download("SPY AAPL", period="1mo")
df
```

```python
df = yf.download(' '.join(sp500), period="10y")
df
```

```python
df['Adj Close'].dropna(how= "any", axis=1).to_csv('sp500.csv')
```
